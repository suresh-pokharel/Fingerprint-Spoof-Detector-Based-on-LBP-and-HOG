```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt
from skimage.feature import local_binary_pattern
```

```python
# reading the image
img = imread("C:\\Image_Analysis_CV\\Image_Analysis_CV\\images\\cat.jpg")
plt.axis("off")
plt.imshow(img)
print(img.shape)
```

```python
# resizing image
resized_img = resize(img, (128*4, 64*4))
plt.axis("off")
plt.imshow(resized_img)
print(resized_img.shape)
```

```python
#creating hog features
fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8),
                	cells_per_block=(2, 2), visualize=True, multichannel=True)
plt.axis("off")
plt.imshow(hog_image, cmap="gray")
```

```python
type(fd)
fd.size
```

```python
# save the images
#plt.imsave("resized_img.jpg", resized_img)
plt.imsave("C:\\Image_Analysis_CV\\Image_Analysis_CV\\images\\hog_image.jpg", hog_image, cmap="gray")
```

```python
radius = 3
n_points = 8 * radius
METHOD = 'uniform'
from skimage.color import rgb2gray
gray_img = rgb2gray(resized_img)
lbp = local_binary_pattern(gray_img, n_points, radius, METHOD)
type(lbp)
lbp.size
import scipy


histogram=scipy.stats.itemfreq(lbp)
type(histogram)
histogram.size
```
